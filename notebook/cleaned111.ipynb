```python
df1.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df1['Classes']=df1['Classes'].str.strip()
```

```python
X = df1.drop('FWI',axis=1)
y = df1['FWI']
y
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
X_train.shape, X_test.shape
```

```python
X_train.corr()
```

```python
def correlation(dataset, Threshold):
    coll = set()
    corr_matrix= dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>Threshold:
                col_name=corr_matrix.columns[i]
                coll.add(col_name)
    return coll
```

```python
corr_features = correlation(X_train,0.85)
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace = True)
X_test.drop(corr_features,axis=1,inplace = True)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lregression = LinearRegression()
lregression.fit(X_train,y_train)
y_pred = lregression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
l1regression = Lasso()
l1regression.fit(X_train,y_train)
y_pred = l1regression.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
l1regressioncv = LassoCV(cv=5)
l1regressioncv.fit(X_train,y_train)
y_pred = l1regressioncv.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
l2regression = Ridge()
l2regression.fit(X_train,y_train)
y_pred = l2regression.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
l2regressioncv = RidgeCV(cv=5)
l2regressioncv.fit(X_train,y_train)
y_pred = l2regressioncv.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
elastic = ElasticNet()
elastic.fit(X_train,y_train)
y_pred = elastic.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
elasticcv = ElasticNetCV(cv=5)
elasticcv.fit(X_train,y_train)
y_pred = elasticcv.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print(mae, mse, score)
```
